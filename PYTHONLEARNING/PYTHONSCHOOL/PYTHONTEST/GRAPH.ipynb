```python
import matplotlib.pyplot as plt

# กำหนดค่าให้กับแกน X และ Y
Q = [0, 50, 100]
S = [10, 30, 50]
S1 = [30, 50, 70]
D = [70, 50, 30]

# วาดกราฟ
plt.figure(figsize=(10, 6))
plt.plot(Q, S, label='อุปทานเดิม (S)', color='black')
plt.plot(Q, S1, label='อุปทานใหม่ (S1)', color='red', linestyle='dashed')
plt.plot(Q, D, label='อุปสงค์ (D)', color='blue')
plt.axhline(y=30, color='gray', linestyle='dotted')
plt.axvline(x=50, color='gray', linestyle='dotted')

# เพิ่มจุดดุลยภาพ
plt.scatter(50, 30, color='black')
plt.text(50, 30, ' E', fontsize=12, verticalalignment='bottom', horizontalalignment='right', color='black')
plt.scatter(35, 45, color='red')
plt.text(35, 45, ' E1', fontsize=12, verticalalignment='bottom', horizontalalignment='right', color='red')

# กำหนดชื่อแกนและกราฟ
plt.xlabel('ปริมาณ (Q)')
plt.ylabel('ราคา (P)')
plt.title('กราฟอุปสงค์และอุปทานของน้ำมัน')
plt.legend()
plt.grid(True)

# แสดงกราฟ
plt.show()
```
